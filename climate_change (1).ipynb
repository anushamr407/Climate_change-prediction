```python
import numpy as np 
import pandas as pd # data processing
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
#to read the input data
df = pd.read_csv('C:/Users/Anusha/Downloads/FAOSTAT_data_1-10-2022.csv/FAOSTAT_data_1-10-2022.csv',encoding='cp1252')
#to print five rows
df.head()
```

```python
#to remove all unwanted  columns
df_new = df.drop(columns = ['Domain','ï»¿Domain Code',  'Element Code', 'Element', 'Months', 'Year Code', 'Unit','Flag','Flag Description'])
df.drop(df.columns[[0]],axis = 1)
```

```python
##PLOTS FOR TRAINING DATA
```

```python

#Plotting graphs for change in mean surface temperature for the following =countries.
# India
India_df = df_new[df_new['Area Code (FAO)'] ==  100]
title_value = India_df['Area'].iat[1]
fig = India_df.plot.scatter(x='Year',y='Value', title=title_value)
fig.set_ylabel('Mean surface temperature change(°C)')
```

```python
# USA
USA_df = df_new[df_new['Area Code (FAO)'] ==  231]
title_value = USA_df['Area'].iat[1]
fig = USA_df.plot.scatter(x='Year',y='Value', title=title_value)
fig.set_ylabel('Mean surface temperature change(°C)')
```

```python
# Singapore
Singapore_df = df_new[df_new['Area Code (FAO)'] ==  200]
title_value = Singapore_df['Area'].iat[1]
fig = Singapore_df.plot.scatter(x='Year',y='Value', title=title_value)
fig.set_ylabel('Mean surface temperature change(°C)')
```

```python
# United Arab Emirates
UAE_df = df_new[df_new['Area Code (FAO)'] ==  225]
title_value = UAE_df['Area'].iat[1]
fig = UAE_df.plot.scatter(x='Year',y='Value', title=title_value)
fig.set_ylabel('Mean surface temperature change(°C)')
```

```python
# Canada
Canada_df = df_new[df_new['Area Code (FAO)'] ==  33]
title_value = Canada_df['Area'].iat[1]
fig = Canada_df.plot.scatter(x='Year',y='Value', title=title_value)
fig.set_ylabel('Mean surface temperature change(°C)')
```

```python
#Removing all the empty records before performing Linear Regression

Singapore_df = Singapore_df.dropna()
USA_df = USA_df.dropna()
Canada_df = Canada_df.dropna()
UAE_df = UAE_df.dropna()
India_df = India_df.dropna()
```

```python
#plots for prediction using LINEAR REGRESSION MODEL
```

```python
# Performing Linear Regression to predict the future Mean Surface Temperature Change for countries
# USA
X = USA_df.iloc[:, 3].values.reshape(-1, 1)
Y = USA_df.iloc[:, 4].values.reshape(-1, 1)
lr = LinearRegression()
lr.fit(X, Y)
Y_pred = lr.predict(X)
year=2050
print(f'In the year {year}, The predicted mean surface temperature change for USA would be {lr.predict([[year]])[0][0]}°C')

plt.scatter(X, Y)
plt.title('USA')
plt.ylabel('Mean surface temperature change(°C)')
plt.xlabel('Year')
plt.ylim(-4,4)
plt.plot(X, Y_pred, color='red')
plt.show()
```
